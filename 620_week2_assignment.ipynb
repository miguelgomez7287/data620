```python
import networkx as nx
import matplotlib.pyplot as plt

# Create the graph
G1 = nx.Graph()

# Define the nodes
nds = ["Andre", "Beverly", "Carol", "Diane", "Ed", "Fernando", 
         "Garth", "Heather", "Ike", "Jane"]

G1.add_nodes_from(nds)

# Define the edges from the Krackhardt kite graph example
egs = [
    ("Andre", "Beverly"), ("Andre", "Carol"), ("Andre", "Diane"),
    ("Beverly", "Diane"), ("Beverly", "Ed"),
    ("Carol", "Diane"), ("Carol", "Fernando"),
    ("Diane", "Fernando"), ("Diane", "Garth"),
    ("Ed", "Garth"),
    ("Fernando", "Garth"), ("Fernando", "Heather"),
    ("Garth", "Heather"),
    ("Heather", "Ike"),
    ("Ike", "Jane")
]

G1.add_edges_from(egs)

# Draw the graph
plt.figure(figsize=(8,6))
nx.draw(G1, with_labels=True, node_color="lightblue", node_size=1500, font_size=10, font_weight="bold")
plt.show()
```
